```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
scores_df= pd.read_csv("wallet_scores.csv")
# df.head()
```

```python
summary = scores_df.describe()
summary 
```

```python
# Calculate top and bottom scoring wallets
top_wallets = scores_df.sort_values(by='score', ascending=False).head(5)
top_wallets
```

```python
bottom_wallets = scores_df.sort_values(by='score', ascending=True).head(5)
bottom_wallets
```

```python

plt.figure(figsize=(10, 5))
sns.histplot(scores_df['score'], bins=30, kde=True, color='#4CAF50', edgecolor='black', linewidth=0.6)
plt.title('Wallet Credit Score Distribution', fontsize=14, fontweight='bold', color='#333333')
plt.xlabel('Credit Score', fontsize=12, fontweight='bold')
plt.ylabel('Number of Wallets', fontsize=12, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.savefig("score_distribution_colored.png")
plt.show()
```
