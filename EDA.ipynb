✅ More deposit → Positive impact on score.

✅ More repay → Positive impact.

🚫 More borrow → Neutral or slightly negative impact.

🚫 More redeemunderlying → Neutral or slight negative impact (depends on strategy).

🚨 More liquidationcall → Strong negative impact.

calculating scores

```python
weights = {
    'deposit': 2.0,
    'repay': 1.5,
    'borrow': -1.0,
    'redeemunderlying': -0.5,
    'liquidationcall': -3.0
}

def calculate_score(row):
    score = 0
    for action in weights:
        score += row.get(action, 0) * weights[action]
    return score

action_counts['raw_score'] = action_counts.apply(calculate_score, axis=1)
```
